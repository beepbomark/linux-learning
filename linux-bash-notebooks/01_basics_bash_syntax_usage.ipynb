## 01. Basic Bash Syntax and Usage

### 1. Understanding basic commands

#### 1.1. Echo Command Basics:
Write a Bash script that uses the echo command to print "Hello, World!" to the terminal.

```bash
#!/bin/bash 

echo "Hello, World!"
```

* #!/bin/bash - called a shebang line, tells the system which interpreter to use for executing the script.
* echo "Hello, World!" - echo command is used to display messages or variables on the terminal.

#### 1.2. Echo with Variables:
Write a bash script that takes a user's name as input and uses the echo command to greet the user with "Hello, [name]!".

```bash
#!/bin/bash

echo "Input your name:"

#read name

name="Hanlong"

echo "Hello, ${name}"
```

* The `read` command accepts user input from standard input.
* `name="Hanlong"` - Assigns a string value to the variable `name`.

### 5. Looping

#### 5.1 Write a Bash script that uses a for loop to print numbers from 0 to 10.

```bash
#!/bin/bash

for i in {0..10}; do
    echo ${i}
done
```

#### 5.2 Write a Bash script that utilizes a while loop to count down from 10 to 1 and then prints "Bash Script!".

```bash
#!/bin/bash

count=10

while (( count > 0 )); do
    echo "${count}"
    (( count-- ))
    sleep 1
done

echo "Bash Script!"
```

#### 5.3 Write a Bash script that prompts the user to input a number and then uses a for loop to print the multiplication table of that number up to 10.

```bash
#!/bin/bash

echo "Enter a number:"
# read num

num=2

for i in {1..10}; do
    echo "${num} x ${i} = $(( num * i ))"
done
```
