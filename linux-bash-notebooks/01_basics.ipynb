## A. Basic Bash Syntax and Usage

### 1. Understanding basic commands

#### 1.1. Echo Command Basics:
Write a Bash script that uses the echo command to print "Hello, World!" to the terminal.

```bash
#!/bin/bash 

echo "Hello, World!"
```

* #!/bin/bash - called a shebang line, tells the system which interpreter to use for executing the script.
* echo "Hello, World!" - echo command is used to display messages or variables on the terminal.

#### 1.2. Echo with Variables:
Write a bash script that takes a user's name as input and uses the echo command to greet the user with "Hello, [name]!".

```bash
#!/bin/bash

echo "Input your name:"

#read name

name="Hanlong"

echo "Hello, ${name}"
```

* The `read` command accepts user input from standard input.
* `name="Hanlong"` - Assigns a string value to the variable `name`.

### 10. Echo with Formatting:

Write a Bash script that uses the echo command to display a message with formatted text, such as bold or colored text, using ANSI escape codes.

```bash
#!/bin/bash

# Bold text
echo -e "\e[1mThis text is bold\e[0m"

# Colored text
echo -e "\e[31mThis text is red\e[0m"
echo -e "\e[32mThis text is green\e[0m"
echo -e "\e[34mThis text is blue\e[0m"

# Bold + colored text
echo -e "\e[1;33mBold yellow text\e[0m"
```

#### 2.10 Using Special Variables:
